```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PowerTransformer
```

```python
data = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
data
```

```python
#1

numericals = data.select_dtypes([np.number])
numericals
```

```python
#2

#variables that check the normal distribution:
    # None
    
#variables that check the exponential distribution:
    # Customer Lifetime Value
    # Income
    # Monthly premium auto
    # Number of Open Complaints
    

cols = list(numericals.columns)

for cols in numericals:
    sns.displot(data = numericals, x = cols, height = 6)
    plt.xticks(rotation = 90)
    plt.show()
```

```python
#3

#USING STANDARD SCALER

numericals1 = numericals.copy()

transformer1 = StandardScaler()
transformer1.fit(numericals1)
numericals_standscale = transformer1.transform(numericals1)
numericals_standscale = pd.DataFrame(numericals_standscale, columns = numericals.columns)

cols = list(numericals_standscale.columns)

for cols in numericals_standscale:
    sns.displot(data = numericals_standscale, x = cols, height = 6)
    plt.xticks(rotation = 90)
    plt.show()

#CONCLUSION: Standard sclarer don't change the distribution shape. It only changed the values scaling.


```

```python
#USING MINMAX SCALER

numericals2 = numericals.copy()

transformer2 = MinMaxScaler()
transformer2.fit(numericals2)
numericals_minmax = transformer2.transform(numericals2)
numericals_minmax = pd.DataFrame(numericals_minmax, columns = numericals.columns)

cols = list(numericals_minmax.columns)

for cols in numericals_minmax:
    sns.displot(data = numericals_minmax, x = cols, height = 6)
    plt.xticks(rotation = 90)
    plt.show()

#CONCLUSION: Min Max Scaler don't change the distribution shape. It only changed the values scaling, this time for a range betweem 0 and 1.
```

```python
#USING POWER TRANSFORMER

numericals3 = numericals.copy()

transformer3 = PowerTransformer()
transformer3.fit(numericals3)
numericals_power = transformer3.transform(numericals3)
numericals_power = pd.DataFrame(numericals_power, columns = numericals.columns)

cols = list(numericals_power.columns)

for cols in numericals_power:
    sns.displot(data = numericals_power, x = cols, height = 6)
    plt.xticks(rotation = 90)
    plt.show()

#CONCLUSION: Power transform give us very different results:
    # We don't have skwed distributions
    # some of the distribution turn to distributions that are more close to a normal one, this is the case of 
        #'Customer Life Time' and 'Total Claim Amount', 'Monts since Last Claim' variables
    # Variables with less unique values, so the discrete ones, got strange results. For example
        # the variable 'Number of open complains' onbly have now 2 unique values.
    
```

```python
#USING LOG TRANSFORMATION

#OPTION 1 with infinte values as 0

def log_transform_clean1(x):
    x = np.log10(x) 
    if np.isfinite(x):
        return x
    else:
        return 0
    
numericals4 = numericals.copy()

cols = list(numericals.columns)
new_list=[]


for cols in numericals4:
    numericals4[cols] = numericals4[cols].apply(log_transform_clean1)

numericals_log = numericals4

for cols in numericals_log:
    sns.displot(data = numericals_log, x = cols, height = 6)
    plt.xticks(rotation = 90)
    plt.show()

#CONCLUSION: With log transformation option 1 we are not getting normal distributions for this data set. In fact some of them that were 
    #left skewed are now righ skewed distribuited. 
```

```python
#OPTION  with infinte values as Nan

def log_transform_clean2(x):
    x = np.log10(x) 
    if np.isfinite(x):
        return x
    else:
        return np.NAN
    
numericals5 = numericals.copy()

cols = list(numericals.columns)
new_list=[]


for cols in numericals5:
    numericals5[cols] = numericals5[cols].apply(log_transform_clean2)

numericals_log2 = numericals5

for cols in numericals_log2:
    numericals_log2[cols] = numericals_log2[cols].fillna(numericals_log2[cols].mean())

#numericals_log2.info()

for cols in numericals_log2:
    sns.displot(data = numericals_log2, x = cols, height = 6)
    plt.xticks(rotation = 90)
    plt.show()

#CONCLUSION: With log transformation option 2  we are not gettin better results comparing to log option 1, since we still got
    # skewed distributions and we get new peaks in some variables since we fill nan with the mean
```
